```python
import pandas as pd
import plotly.express as px
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import matplotlib.patches as mpatches


df = pd.read_csv('diabetes_dataset.csv')
data= df.copy()
```

## Data Cleaning and Understanding

```python
data.head()
data.info()
data.describe()
```

```python
print(data.isnull().sum())
print(f'Duplicated:',data.duplicated().sum())
```

**Explanation:** I decided to drop the unknown column as it provided no relevance in the data analysis. There were no duplicated values, but there is a high number of missing values(3,320) in the Alcohol Consumption column. Due to this high number of missing values, I will do data imputation. 

```python
data = data.drop('Unnamed: 0', axis=1)
alcohol_mode = data['Alcohol_Consumption'].mode()[0]
data['Alcohol_Consumption'].fillna(alcohol_mode, inplace=True)
```

**Explanation:** Due to the high number of missing values, dropping the Alcohol Consumption Cateogory would lead to a substantial loss of information. This could potentially produce selection bias. The cateogorical nature of this column led me to do a mode imputation to fill in those missing values. The mode was moderate,so by imputing the mode I can preserve the sample size and maintaint the overall distribution of data. 

```python
black_alcohol_consumers = data[(data['Ethnicity'] == 'Black') & (data['Alcohol_Consumption'] != 'None')]

print(f"Number of Black Alcohol Consumers: {len(black_alcohol_consumers)}")
black_alcohol_consumers.head()
```

**Explanation:** To look at specific groups of people, I can filter the data to show me people who are a part of a particular group. In this example, I filtered to look at black participants who reported consuming moderate to heavy amounts of alcohol. This allows me to see if there is a relationship between alcohol consumption and ethnicity, while also looking at other sections of data like BMI.

```python
bmi_bins=[0,18.5,25.0,30.0, float('inf')]#based on standard health ranges
bmi_labels= ['Underweight','Normal Weight', 'Overweight', 'Obese']
data['BMI_Category'] = pd.cut(
    data['BMI'],
    bins=bmi_bins,
    labels=bmi_labels,
    right=False
)
print("BMI Category Counts:")
print(data['BMI_Category'].value_counts())

glucose_bins= [0, 5.7, 6.5, float('inf')]
glucose_labels=['Normal', 'Pre-diabetes', 'Diabetes']
data['Glucose_Control_Status'] = pd.cut(
    data['HbA1c'],
    bins= glucose_bins,
    labels= glucose_labels,
    right= False
)
print("\n GLucose Control Status")
print(data["Glucose_Control_Status"].value_counts())

age_bins = [0, 30, 50, 65, float('inf')]
age_labels = ['Young Adult', 'Middle-Aged', 'Senior', 'Elderly']
data['Age_Group'] = pd.cut(
    data['Age'],
    bins=age_bins,
    labels=age_labels,
    right=False
)
print("\n Age Category")
print(data['Age_Group'].value_counts())
```

**Explanation:** I createed a new categorical column for BMI, Age, and glucose to assist in future data visualization, and to also see the difference in populations. I was able to find sources that provided the parameters for each category([NHF Inform:BMI](https://www.nhsinform.scot/healthy-living/weight-loss/understanding-healthy-weight/how-do-i-measure-my-body-mass-index-bmi))([CDC: glucose](https://www.cdc.gov/diabetes/diabetes-testing/prediabetes-a1c-test.html#:~:text=A1C%20results,Diabetes%3A%206.5%25%20or%20above)).

```python
categorical_cols1 = ['Sex', 'Ethnicity', 'Physical_Activity_Level', 'Smoking_Status', 'Alcohol_Consumption', 'Family_History_of_Diabetes', 'Previous_Gestational_Diabetes']

for col in categorical_cols1:
        data[col] = data[col].astype('category')

numeric_cols = [
    'Age', 'BMI', 'Waist_Circumference', 'Fasting_Blood_Glucose', 'HbA1c',
    'Blood_Pressure_Systolic', 'Blood_Pressure_Diastolic',
    'Cholesterol_Total', 'Cholesterol_HDL', 'Cholesterol_LDL',
    'GGT', 'Serum_Urate'
]
for col in numeric_cols:
    data[col] = pd.to_numeric(data[col], errors='coerce')

for col in numeric_cols:
    if data[col].isnull().any():
        data[col].fillna(data[col].mean(), inplace=True)
```

```python
print("Summary Statistics of Continuous Variables")
print(data.describe().T)
print("\n Unique Values for Categorical Variables")
categorical_cols2 = ['Sex', 'Ethnicity', 'Physical_Activity_Level', 
                    'Alcohol_Consumption', 'Smoking_Status', 
                    'Family_History_of_Diabetes', 'Previous_Gestational_Diabetes',
                    'BMI_Category', 'Glucose_Control_Status', 'Age_Group']

for col in categorical_cols2:
    print(f"\n{col}:")
    print(data[col].value_counts())
```

## Data Visualization

### Heatmap

```python
numeric_cols = [
    'Age', 'BMI', 'Waist_Circumference', 'Fasting_Blood_Glucose', 'HbA1c',
    'Blood_Pressure_Systolic', 'Blood_Pressure_Diastolic',
    'Cholesterol_Total', 'Cholesterol_HDL', 'Cholesterol_LDL',
    'GGT', 'Serum_Urate'
]
correlation_matrix = data[numeric_cols].corr() 

plt.figure(figsize = (12, 10))
sns.heatmap(
    correlation_matrix,
    annot = True,
    fmt = ".2f",
    cmap = 'coolwarm',
    linewidths = .5,
    cbar_kws = {'label': 'Correlation Coefficient'}
)
plt.title('Heatmap for Diabetes Dataset', fontsize = 16)
plt.xticks(rotation = 45, ha = 'right')
plt.yticks(rotation = 0)
plt.tight_layout()
plt.savefig('heatmap_correlation_matrix.png')
plt.show()
```

**Interpretation:** This visualization shows a highly balanced distribution of possible hereditary risk across the entire population sample.  I say this because the number of individuals with a family history(5070) is almost equal to those without(4930). When separated by gender, the plot reveals that among those who don't have a family history, females slightly outnumber males, but it is the opposite for those who do have a family history. This suggests that while hereditary risks are prevalent, the male subgroup in this data shows a slightly greater predisposition based on family history.

### Fasting Blood Glucose: Histogram

```python

conditions = [
    (data['Fasting_Blood_Glucose'] < 100),
    (data['Fasting_Blood_Glucose'] >= 100) & (data['Fasting_Blood_Glucose'] < 126),
    (data['Fasting_Blood_Glucose'] >= 126)
]

choices = [
    'Normal Range (< 100)',
    'Pre-Diabetes Risk (100-125)',
    'Diabetes (>= 126)'
]

data['FBG_3_Risk_Level'] = np.select(conditions, choices, default='Error')

data['FBG_3_Risk_Level'] = data['FBG_3_Risk_Level'].astype('category')

print("3-Level FBG Risk Distribution:")
print(data['FBG_3_Risk_Level'].value_counts())
```

```python
import matplotlib.patches as mpatches

plt.figure(figsize=(10, 6))

custom_palette = {
    'Normal Range (< 100)': 'green',
    'Pre-Diabetes Risk (100-125)': 'orange',
    'Diabetes (>= 126)': 'red'
}

sns.histplot(
    data=data,
    x='Fasting_Blood_Glucose',
    bins=30,
    hue='FBG_3_Risk_Level',
    multiple='stack',
    palette=custom_palette,
    kde=False
)


import matplotlib.patches as mpatches

handles = [mpatches.Patch(color=custom_palette[label], label=label) for label in custom_palette]

plt.legend(
    handles=handles,
    title='FBG Risk Status',
    loc='lower right'
)

plt.title('Distribution of FBG Levels by 3 Clinical Risk Categories', fontsize=16)
plt.xlabel('Fasting Blood Glucose (mg/dL)', fontsize=12)
plt.ylabel('Count of Individuals', fontsize=12)
plt.savefig('FBG_dis_histogram.png')
plt.show()
```

**Intepretation:** The histogram provides a visual distribution of Blood Glucose Levels which provides a description of the health profile for this population. The overall distribution is skewed mostly to the right with there being more individuals with high risk FBG levels. The different colors allows me to see where there is a higher concentration of idnividuals depending on the FBG status. The parameters for each status was based around information found online([MayoClinic](https://www.mayoclinic.org/diseases-conditions/diabetes/diagnosis-treatment/drc-20371451#:~:text=A%20fasting%20blood%20sugar%20level,Glucose%20tolerance%20test.)). Majority of particpants are living with pre-diabetes or diabetes based on FBG levels. The data shows that most of this population suffers from impaired glucose control. 

### HbA1C by Activity Level: Violin Plot

```python
NORMAL_MAX = 5.7
DIABETES_MIN = 6.5
DATA_MAX = data['HbA1c'].max()
plt.figure(figsize=(10, 6))
plt.axhspan(0, NORMAL_MAX, color='green', alpha=0.1, label='Normal Range')
plt.axhspan(DIABETES_MIN, DATA_MAX + 0.5, color='red', alpha=0.15, label='Diabetes Range')

sns.violinplot(
    data=data,
    x='Physical_Activity_Level',
    y='HbA1c',
    order=['Low', 'Moderate', 'High'], 
    palette=['gold','orange','maroon'] 
)


plt.title('HbA1c Distribution by Physical Activity Level', fontsize=16)
plt.xlabel('Physical Activity Level', fontsize=12)
plt.ylabel('HbA1c (%)', fontsize=12)
plt.legend(loc='lower right')
plt.savefig('hbA1c_activity_violinplot.png')
plt.show()
```

**Interpretation:** This visualization reveals a clear and expected relationship between lifestyle and blood glucose control. HbA1c values are clustered across all three activity levels, but there are distinct differences in distribution. All activity levels share a similar median, which is above the diabetic threshold. However, the plot suggests that the low activity group has the widest distribution. The low activity group has a higher density within the upper range, which indicates that those who reported low activity levels have poorer glucose control when compared to other groups. The overall differences are not drastically different.

### BMI v. Waist Circumference: Box Plot

```python
r_value = data['BMI'].corr(data['Waist_Circumference'])

plt.figure(figsize=(8, 8))

sns.scatterplot(
    data=data,
    x='BMI',
    y='Waist_Circumference',
    alpha=0.2, 
    s=10, 
    color='red'
)

sns.regplot(
    data=data,
    x='BMI',
    y='Waist_Circumference',
    scatter=False,
    color='red',
    line_kws={'linestyle': '-', 'linewidth': 2}
)

annotation_text = f"Correlation (r) = {r_value:.2f}"

plt.text(
    x=20, 
    y=115, 
    s=annotation_text, 
    fontsize=14, 
    color='black', 
    bbox=dict(facecolor='white', alpha=0.7, edgecolor='none', boxstyle='round,pad=0.5')
)

plt.title('BMI vs. Waist Circumference: Strong Correlation Check', fontsize=16)
plt.xlabel('Body Mass Index (BMI)', fontsize=12)
plt.ylabel('Waist Circumference (cm)', fontsize=12)
plt.grid(True, linestyle=':', alpha=0.5)
plt.savefig('BMI_Waist_scatterplot.png')
plt.show()
```

**Interpretation:** This scatter plot visualizes a relatively weak negative correlation between waist circumference, which suggests that waist circumference is not a good predictor for BMI. 

## Animated Visualizations

### How does the distribution of high-risk glucose categories (Pre-diabetes and Diabetes) evolve across BMI Categories as the population ages? *Using an Animated Stacked Bar Plot*

```python

risk_progression = data.groupby(['Age_Group', 'BMI_Category', 'Glucose_Control_Status']).size().reset_index(name='Count')

age_order = ['Young Adult', 'Middle-Aged', 'Senior', 'Elderly']
bmi_order = ['Underweight', 'Normal Weight', 'Overweight', 'Obese']
glucose_order = ['Normal', 'Pre-diabetes', 'Diabetes'] # Stack order

fig = px.bar(
    risk_progression,
    x="BMI_Category",
    y="Count",
    color="Glucose_Control_Status",
    animation_frame="Age_Group", 
    category_orders={
        "Age_Group": age_order,
        "BMI_Category": bmi_order,
        "Glucose_Control_Status": glucose_order
    },
    title="Evolution of Glucose Risk Across BMI Categories by Age"
)

fig.update_layout(
    xaxis_title="BMI Category",
    yaxis_title="Count of Individuals",
    legend_title="Glucose Status"
)

fig.show()
fig.write_html("BMI_Age_status.html")
```

**Interpretation:** This animated visual shows the progression of BMI and glucose categories across age groups. This provided an insightful view into the intersection of obesity and aging as possible driving factors of diabetes risk. This population mostly consists of middle-aged individuals, and when looking at that age group, the obese group is made up mostly of those with diabetes. Based on this visual, it can be said that middle age and higher BMI could put individuals at a higher risk of diabetes. Overall, the obese group had a higher count of individuals with diabetes across all groups.

### How does the trade-off between the two primary cardiovascular risk factors(Blood Pressure and Cholesterol) differ between individuals with a normal glucose status versus those with diabetes? *Using box plots*

```python

risk_order = ['Normal', 'Pre-diabetes', 'Diabetes']
age_order = ['Young Adult', 'Middle-Aged', 'Senior', 'Elderly']
custom_palette = {
    'Normal': 'green', 
    'Pre-diabetes': 'orange', 
    'Diabetes': 'red'
}


metrics_to_melt = ['Cholesterol_Total', 'Blood_Pressure_Systolic']
data_long = pd.melt(
    data, 
    id_vars=['Age_Group', 'Glucose_Control_Status'], 
    value_vars=metrics_to_melt, 
    var_name='Metric', 
    value_name='Value'
)

fig_combined = px.box(
    data_long,
    x='Age_Group',
    y='Value',
    color='Glucose_Control_Status',
    facet_row='Metric',  
    
    category_orders={
        'Age_Group': age_order, 
        'Glucose_Control_Status': risk_order,
        'Metric': metrics_to_melt
    },
    color_discrete_map=custom_palette,
    points='all', 
    title='Distribution of Cardiovascular Risk Metrics by Age and Glucose Status(Interactive)'
)

fig_combined.update_layout(height=800, showlegend=True)
fig_combined.for_each_annotation(lambda a: a.update(text=a.text.replace('Metric=', ''))) 

fig_combined.show()
fig.write_html("Glucose_cholestorolBP_boxplot.html")
```

**Interpretation:** The three plots demonstrate that there is a severe increase in risk when comparing the normal group to the other two groups across the ages. The diabetes group exhibits a higher median cholesterol and blood pressure than the normal group. The position of the margin boxes indicates that poor glucose regulation increases cardiovascular risk. The median blood pressure doesn't change change across groups regardless of age or glucose control status.
